```python
import tensorflow as tf
from tensorflow.keras import layers, models
import numpy as np
from PIL import Image, ImageFilter
import matplotlib.pyplot as plt
from tkinter import Tk, filedialog

# -----------------------------
# 1. MNIST (НАВЧАЛЬНИЙ ДАТАСЕТ)
# -----------------------------

print("Loading MNIST...")

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

# -----------------------------
# 2. CNN модель
# -----------------------------

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D(2, 2),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# -----------------------------
# 3. НАВЧАННЯ
# -----------------------------

print("Training...")

model.fit(x_train, y_train, epochs=5)

# -----------------------------
# 4. ТЕСТ MNIST
# -----------------------------

loss, acc = model.evaluate(x_test, y_test)
print("MNIST accuracy:", acc)

# -----------------------------
# 5. ПРЕПРОЦЕСИНГ ФОТО
# -----------------------------

def preprocess_image(path):

    img = Image.open(path).convert('L')

    # інверсія
    img = np.array(img)
    img = 255 - img

    # бінаризація
    img = (img > 100) * 255

    # знаходження цифри
    coords = np.column_stack(np.where(img > 0))

    if coords.size == 0:
        return None

    y_min, x_min = coords.min(axis=0)
    y_max, x_max = coords.max(axis=0)

    img = img[y_min:y_max, x_min:x_max]

    img = Image.fromarray(img.astype(np.uint8))

    # ROF-подібна очистка
    img = img.filter(ImageFilter.MedianFilter(3))

    # вписування в квадрат
    max_side = max(img.size)
    new_img = Image.new("L", (max_side, max_side), 0)

    new_img.paste(
        img,
        ((max_side - img.size[0]) // 2,
         (max_side - img.size[1]) // 2)
    )

    # масштабування
    img = new_img.resize((28, 28))

    img = np.array(img) / 255.0

    return img.reshape(1, 28, 28, 1)

# -----------------------------
# 6. РОЗПІЗНАВАННЯ З ОБЕРТАННЯМ
# -----------------------------

def predict_with_rotation(path):

    angles = range(-30, 31, 5)

    best_digit = None
    best_confidence = 0

    original = Image.open(path).convert('L')

    for angle in angles:

        rotated = original.rotate(angle)

        rotated.save("temp.png")

        img = preprocess_image("temp.png")

        if img is None:
            continue

        prediction = model.predict(img)

        digit = np.argmax(prediction)
        confidence = np.max(prediction)

        if confidence > best_confidence:
            best_confidence = confidence
            best_digit = digit

    return best_digit, best_confidence

# -----------------------------
# 7. ВИБІР ФАЙЛУ
# -----------------------------

root = Tk()
root.withdraw()

file_path = filedialog.askopenfilename(
    title="Вибери фото цифри",
    filetypes=[("Image files", "*.png *.jpg *.jpeg *.bmp")]
)

# -----------------------------
# 8. ЗАПУСК
# -----------------------------

if file_path:

    digit, conf = predict_with_rotation(file_path)

    print("Файл:", file_path)
    print("Розпізнано:", digit)
    print("Ймовірність:", conf)

    # показ зображення
    img = Image.open(file_path)
    plt.imshow(img, cmap='gray')
    plt.title(f"Результат: {digit}")
    plt.axis('off')
    plt.show()

else:
    print("Файл не вибрано")
```
