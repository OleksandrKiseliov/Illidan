```python
import random
import math
import matplotlib.pyplot as plt

def random_state(n):
    """Генерує випадковий стан (розташування ферзів у стовпцях)."""
    return [random.randint(0, n - 1) for _ in range(n)]

def conflicts(state):
    """Обчислює кількість конфліктів між ферзями."""
    n = len(state)
    conflicts_count = 0

    for i in range(n):
        for j in range(i + 1, n):
            if state[i] == state[j] or abs(state[i] - state[j]) == abs(i - j):
                conflicts_count += 1

    return conflicts_count

def random_neighbor(state):
    """Генерує сусідній стан, змінюючи позицію випадкового ферзя."""
    neighbor = state[:]
    n = len(state)
    col = random.randint(0, n - 1)
    row = random.randint(0, n - 1)
    neighbor[col] = row
    return neighbor

def simulated_annealing(n, initial_temp=1.0, cooling_rate=0.99, min_temp=0.001):
    """Розв’язує задачу N-ферзів за допомогою алгоритму відпалу."""
    current = random_state(n)
    temp = initial_temp

    while temp > min_temp:
        neighbor = random_neighbor(current)
        delta = conflicts(neighbor) - conflicts(current)

        if delta < 0 or random.random() < math.exp(-delta / temp):
            current = neighbor

        if conflicts(current) == 0:
            return current

        temp *= cooling_rate

    return None

def print_solution(state):
    """Виводить дошку з розташуванням ферзів у текстовому форматі."""
    if state is None:
        print("Рішення не знайдено")
        return

    n = len(state)
    for row in range(n):
        line = ''.join('Q' if state[col] == row else '.' for col in range(n))
        print(line)

def visualize_solution(state):
    """Візуалізує розташування ферзів на шаховій дошці."""
    if state is None:
        print("Рішення не знайдено")
        return

    n = len(state)
    fig, ax = plt.subplots(figsize=(n, n))
    ax.set_xlim(-0.5, n - 0.5)
    ax.set_ylim(-0.5, n - 0.5)

    # Малюємо шахову дошку
    for i in range(n):
        for j in range(n):
            color = 'lightgray' if (i + j) % 2 == 0 else 'white'
            ax.add_patch(plt.Rectangle((j, i), 1, 1, color=color, ec='black'))

    # Малюємо ферзів
    for col, row in enumerate(state):
        ax.text(col + 0.5, row + 0.5, '\u265B', fontsize=30, ha='center', va='center', color='black')

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_frame_on(False)
    plt.gca().invert_yaxis()  # Інвертуємо вісь Y, щоб верхній рядок відповідав верхній частині шахової дошки
    plt.tight_layout(pad=0.5)  # Покращене масштабування для уникнення обрізаних країв
    plt.show()

# Виконання алгоритму
n = 8  # Задайте розмірність дошки
solution = simulated_annealing(n)
print_solution(solution)
visualize_solution(solution)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# 1. Тестова функція (аналітично інтегровна)
def test_function(x):
    return np.x^2

# 2. Основна функція (аналітично не інтегровна)
def main_function(x):
    return np.exp(x**2)

# 3. Обчислення точного значення інтеграла тестової функції
def exact_integral_test(a, b):
    result, _ = quad(test_function, a, b)
    return result

# 4. Генерування випадкової точки
def random_point(a, b, max_y):
    x = np.random.uniform(a, b)
    y = np.random.uniform(0, max_y)
    return x, y

# 5. Допоміжна функція для обчислення значення функцій в заданій точці
def compute_function_value(x, mode='test'):
    if mode == 'test':
        return test_function(x)
    elif mode == 'main':
        return main_function(x)

# 6. Реалізація методу Монте-Карло
def monte_carlo_integration(func_mode, a, b, points=10000):
    max_y = max(compute_function_value(x, func_mode) for x in np.linspace(a, b, 1000))

    under_curve = 0
    for _ in range(points):
        x_rand, y_rand = random_point(a, b, max_y)
        if y_rand <= compute_function_value(x_rand, func_mode):
            under_curve += 1

    rect_area = (b - a) * max_y
    integral_estimate = (under_curve / points) * rect_area

    return integral_estimate

# 7. Виконання обчислень та оцінка похибки для тестової задачі
a, b = 0, 1
exact_value = exact_integral_test(a, b)
monte_carlo_value = monte_carlo_integration('test', a, b)
absolute_error = abs(exact_value - monte_carlo_value)
relative_error = absolute_error / abs(exact_value)

print(f"Тестова задача (sin(x) на [0, 1]):")
print(f"Точне значення: {exact_value:.6f}")
print(f"Монте-Карло значення: {monte_carlo_value:.6f}")
print(f"Абсолютна похибка: {absolute_error:.6f}")
print(f"Відносна похибка: {relative_error:.6f}")

# 8. Обчислення основної задачі (exp(x^2) на [1, 2])
main_a, main_b = 1, 2
main_integral_estimate = monte_carlo_integration('main', main_a, main_b)
print(f"\nОсновна задача (exp(x^2) на [1, 2]):")
print(f"Оцінка значення інтеграла (Монте-Карло): {main_integral_estimate:.6f}")

# 9. Візуалізація результатів
x_vals_test = np.linspace(a, b, 500)
y_vals_test = [test_function(x) for x in x_vals_test]

x_vals_main = np.linspace(main_a, main_b, 500)
y_vals_main = [main_function(x) for x in x_vals_main]

plt.figure(figsize=(12, 6))

# Графік для тестової функції
plt.subplot(1, 2, 1)
plt.plot(x_vals_test, y_vals_test, label='sin(x)', color='blue')
plt.fill_between(x_vals_test, y_vals_test, alpha=0.3, color='blue')
plt.title('Метод Монте-Карло для sin(x) на [0, 1]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

# Графік для основної функції
plt.subplot(1, 2, 2)
plt.plot(x_vals_main, y_vals_main, label='exp(x^2)', color='green')
plt.fill_between(x_vals_main, y_vals_main, alpha=0.3, color='green')
plt.title('Метод Монте-Карло для exp(x^2) на [1, 2]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, AgglomerativeClustering
from scipy.spatial.distance import cdist

# 1. Генерація даних
N = 1000
np.random.seed(42)
data = np.random.rand(N, 2)

# 2. Міра віддалі (Евклідова)
def euclidean_distance(a, b):
    return np.linalg.norm(a - b)

# 3. Алгоритм кластеризації K-середніх
def kmeans_clustering(data, n_clusters=5):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    labels = kmeans.fit_predict(data)
    centers = kmeans.cluster_centers_
    return labels, centers

# 4. Алгоритм кластеризації ієрархічний
def hierarchical_clustering(data, n_clusters=5):
    hc = AgglomerativeClustering(n_clusters=n_clusters)
    labels = hc.fit_predict(data)
    return labels

# 5. Оцінка якості кластеризації
def average_distance(data, labels, centers):
    total_distance = 0
    count = 0
    for idx, center in enumerate(centers):
        cluster_points = data[labels == idx]
        distances = cdist(cluster_points, [center], metric='euclidean')
        total_distance += distances.sum()
        count += len(cluster_points)
    return total_distance / count

# 6. Кластеризація
k_labels, k_centers = kmeans_clustering(data)
h_labels = hierarchical_clustering(data)

# Порівняння
k_avg_distance = average_distance(data, k_labels, k_centers)

print(f"K-середніх: Середня відстань до центра: {k_avg_distance:.4f}")
print(f"Ієрархічна: Кількість кластерів: {len(np.unique(h_labels))}")

# 7. Візуалізація результатів
plt.figure(figsize=(12, 6))

# K-середніх
plt.subplot(1, 2, 1)
plt.scatter(data[:, 0], data[:, 1], c=k_labels, cmap='tab10', s=10)
plt.scatter(k_centers[:, 0], k_centers[:, 1], c='black', marker='x', label='Centers')
plt.title('K-середніх')
plt.legend()

# Ієрархічна
plt.subplot(1, 2, 2)
plt.scatter(data[:, 0], data[:, 1], c=h_labels, cmap='tab10', s=10)
plt.title('Ієрархічна кластеризація')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import random
import pickle
import matplotlib.pyplot as plt

# 1. Генерація карти міст
def generate_map():
    N = random.randint(25, 35)
    distances = np.random.randint(10, 100, size=(N, N))
    distances = (distances + distances.T) // 2  # симетрична матриця
    np.fill_diagonal(distances, 1)  # уникнути нуля в діагоналі (буде видно як близькість до себе)
    return distances

def save_map(distances, filename="map.pkl"):
    with open(filename, 'wb') as f:
        pickle.dump(distances, f)

def load_map(filename="map.pkl"):
    with open(filename, 'rb') as f:
        distances = pickle.load(f)
    return distances

# 2. Мурашиний алгоритм
class AntColony:
    def __init__(self, distances, n_ants, n_best, n_iterations, decay, alpha=1, beta=1):
        self.distances = distances
        self.pheromone = np.ones(self.distances.shape) / len(distances)
        self.all_inds = range(len(distances))
        self.n_ants = n_ants
        self.n_best = n_best
        self.n_iterations = n_iterations
        self.decay = decay
        self.alpha = alpha
        self.beta = beta
        self.history = []

    def run(self):
        shortest_path = None
        all_time_shortest_path = (None, np.inf)
        for i in range(self.n_iterations):
            all_paths = self.gen_all_paths()
            self.spread_pheromone(all_paths, self.n_best)
            shortest_path = min(all_paths, key=lambda x: x[1])
            self.history.append(shortest_path[1])
            if shortest_path[1] < all_time_shortest_path[1]:
                all_time_shortest_path = shortest_path
            self.pheromone *= (1 - self.decay)
        return all_time_shortest_path

    def gen_path_dist(self, path):
        total_dist = 0
        for i in range(len(path) - 1):
            total_dist += self.distances[path[i]][path[i + 1]]
        total_dist += self.distances[path[-1]][path[0]]
        return total_dist

    def gen_all_paths(self):
        all_paths = []
        for i in range(self.n_ants):
            path = self.gen_path(0)
            all_paths.append((path, self.gen_path_dist(path)))
        return all_paths

    def gen_path(self, start):
        path = []
        visited = set()
        visited.add(start)
        path.append(start)
        prev = start
        for i in range(len(self.distances) - 1):
            move = self.pick_move(self.pheromone[prev], self.distances[prev], visited)
            path.append(move)
            visited.add(move)
            prev = move
        return path

    def pick_move(self, pheromone, dist, visited):
        pheromone = np.copy(pheromone)
        pheromone[list(visited)] = 0
        with np.errstate(divide='ignore'):  # придушити попередження
            heuristic = 1.0 / dist
        heuristic[list(visited)] = 0

        row = pheromone ** self.alpha * (heuristic ** self.beta)
        if np.sum(row) == 0:
            row[list(set(range(len(dist))) - visited)] = 1
        norm_row = row / np.sum(row)
        move = np.random.choice(self.all_inds, 1, p=norm_row)[0]
        return move

    def spread_pheromone(self, all_paths, n_best):
        sorted_paths = sorted(all_paths, key=lambda x: x[1])
        for path, dist in sorted_paths[:n_best]:
            for move in range(len(path) - 1):
                self.pheromone[path[move]][path[move + 1]] += 1.0 / self.distances[path[move]][path[move + 1]]
            self.pheromone[path[-1]][path[0]] += 1.0 / self.distances[path[-1]][path[0]]

# 3. Симуляції і візуалізація

def run_simulation_with_visualization(filename="map.pkl", n_ants=50, n_best=5, n_iterations=100, decay=0.5, alpha=1, beta=2):
    distances = load_map(filename)
    colony = AntColony(distances, n_ants, n_best, n_iterations, decay, alpha=alpha, beta=beta)
    best_path, best_length = colony.run()

    print(f"Найкращий шлях: {best_path}")
    print(f"Довжина найкращого шляху: {best_length:.2f}")

    # Створити випадкові координати міст
    np.random.seed(42)
    coordinates = np.random.rand(len(distances), 2)

    # Побудувати графік шляху
    plt.figure(figsize=(8, 8))
    for i in range(len(best_path)):
        start_city = best_path[i]
        end_city = best_path[(i + 1) % len(best_path)]
        plt.plot([coordinates[start_city][0], coordinates[end_city][0]],
                 [coordinates[start_city][1], coordinates[end_city][1]], 'b-')
    plt.scatter(coordinates[:, 0], coordinates[:, 1], c='red')
    for idx, (x, y) in enumerate(coordinates):
        plt.text(x, y, str(idx), fontsize=9, ha='right')
    plt.title('Найкращий знайдений маршрут')
    plt.grid(True)
    plt.show()

    # Побудова графіку зміни довжини маршруту по ітераціях
    plt.figure(figsize=(10, 4))
    plt.plot(colony.history, color='purple')
    plt.xlabel('Ітерація')
    plt.ylabel('Довжина найкращого шляху')
    plt.title('Зміна довжини маршруту протягом ітерацій')
    plt.grid(True)
    plt.show()

# --- Головна логіка ---

# Згенерувати карту
distances = generate_map()
save_map(distances)

# Виконати симуляцію з візуалізацією
run_simulation_with_visualization(n_ants=50, n_best=5, n_iterations=100, decay=0.5, alpha=1, beta=2)
```

```python
import random
import json
from collections import defaultdict
from tabulate import tabulate
import pandas as pd

# 🎓 === Параметри ===
MAX_LESSONS_PER_DAY = 5
POPULATION_SIZE = 50
NUM_GENERATIONS = 100
NUM_CLASSES = 3
NUM_TEACHERS = 4

# 📚 === Предмети ===
lessons = [
    'Математика', 'Українська мова',
    'Біологія', 'Фізкультура',
    'Хімія', 'Музика', 'Хореографія'
]

# 👩‍🏫 === Вчителі і класи ===
teachers = [f'Вчитель_{i + 1}' for i in range(NUM_TEACHERS)]
classes = [f'Клас_{i + 1}' for i in range(NUM_CLASSES)]

# 🏫 === Спеціальні уроки та приміщення ===
special_rooms = {
    'Фізкультура': 'Спортзал',
    'Хореографія': 'Хореозал',
    'Музика': 'Музичний кабінет'
}
available_rooms = classes + list(special_rooms.values())

# 🧠 === Призначення вчителів до предметів ===
subject_teachers = {
    'Математика': 'Вчитель_1',
    'Українська мова': 'Вчитель_2',
    'Біологія': 'Вчитель_3',
    'Фізкультура': 'Вчитель_1',
    'Хімія': 'Вчитель_4',
    'Музика': 'Вчитель_3',
    'Хореографія': 'Вчитель_2'
}

# 🏅 === Класний керівник для кожного класу ===
class_homeroom_teacher = {
    'Клас_1': 'Вчитель_1',
    'Клас_2': 'Вчитель_2',
    'Клас_3': 'Вчитель_3'
}

# 🛠 === Генерація розкладу ===
def generate_schedule():
    schedule = {class_name: {day: {} for day in range(5)} for class_name in classes}
    for day in range(5):
        for class_name in classes:
            used_lessons = random.sample(lessons, MAX_LESSONS_PER_DAY)
            lesson_indices = list(range(MAX_LESSONS_PER_DAY))
            random.shuffle(lesson_indices)
            i = 0
            attempts = 0
            while i < MAX_LESSONS_PER_DAY:
                if attempts > 200:
                    break
                lesson = used_lessons[i]
                teacher = subject_teachers[lesson]
                room = special_rooms.get(lesson, class_name)

                slot_free = True
                for other_class in classes:
                    if other_class == class_name:
                        continue
                    other_lesson = schedule[other_class][day].get(f'Урок_{lesson_indices[i] + 1}')
                    if other_lesson:
                        if other_lesson[1] == teacher or other_lesson[2] == room:
                            slot_free = False
                            break

                if slot_free:
                    schedule[class_name][day][f'Урок_{lesson_indices[i] + 1}'] = (lesson, teacher, room)
                    i += 1
                    attempts = 0
                else:
                    random.shuffle(used_lessons)
                    attempts += 1
    return schedule

# ⚖️ === Фітнес-функція ===
def fitness(schedule):
    penalty = 0
    for day in range(5):
        for i in range(MAX_LESSONS_PER_DAY):
            teachers_set = set()
            rooms_set = set()
            for class_name in classes:
                lesson_info = schedule[class_name][day].get(f'Урок_{i + 1}')
                if lesson_info:
                    lesson, teacher, room = lesson_info
                    if teacher in teachers_set:
                        penalty += 3
                    else:
                        teachers_set.add(teacher)
                    if room in rooms_set:
                        penalty += 2
                    else:
                        rooms_set.add(room)

    for class_name in classes:
        for day in range(5):
            lessons_in_day = [schedule[class_name][day].get(f'Урок_{i + 1}') for i in range(MAX_LESSONS_PER_DAY)]
            has_empty = any(l is None for l in lessons_in_day)
            if has_empty:
                penalty += 5

    for class_name in classes:
        homeroom_teacher = class_homeroom_teacher[class_name]
        homeroom_teacher_lessons = 0
        for day in range(5):
            for i in range(MAX_LESSONS_PER_DAY):
                lesson_info = schedule[class_name][day].get(f'Урок_{i + 1}')
                if lesson_info:
                    if lesson_info[1] == homeroom_teacher:
                        homeroom_teacher_lessons += 1
        if homeroom_teacher_lessons < 13:
            penalty += (13 - homeroom_teacher_lessons) * 10
    return penalty

# 🧬 === Мутація ===
def mutate(schedule):
    new_schedule = {c: {d: schedule[c][d].copy() for d in schedule[c]} for c in schedule}
    class_name = random.choice(classes)
    day = random.randint(0, 4)
    lesson_index = random.randint(0, MAX_LESSONS_PER_DAY - 1)
    lesson_key = f'Урок_{lesson_index + 1}'
    if lesson_key in new_schedule[class_name][day]:
        new_lesson = random.choice(lessons)
        new_teacher = subject_teachers[new_lesson]
        new_room = special_rooms.get(new_lesson, class_name)
        new_schedule[class_name][day][lesson_key] = (new_lesson, new_teacher, new_room)
    return new_schedule

# 🔀 === Схрещування ===
def crossover(sch1, sch2):
    new_schedule = {}
    for class_name in classes:
        new_schedule[class_name] = {}
        for day in range(5):
            source = sch1 if random.random() < 0.5 else sch2
            new_schedule[class_name][day] = source[class_name][day].copy()
    return new_schedule

# 🧪 === Генетичний алгоритм ===
def genetic_algorithm(pop_size, generations):
    population = [generate_schedule() for _ in range(pop_size)]
    for _ in range(generations):
        scored = [(s, fitness(s)) for s in population]
        scored.sort(key=lambda x: x[1])
        top = [x[0] for x in scored[:pop_size // 2]]
        new_population = top.copy()
        while len(new_population) < pop_size:
            if random.random() < 0.5:
                new_population.append(mutate(random.choice(top)))
            else:
                p1, p2 = random.sample(top, 2)
                new_population.append(crossover(p1, p2))
        population = new_population
    best = min(population, key=fitness)
    return best

# 🖼 === Вивід розкладу ===
def print_schedule(schedule):
    days = ["Понеділок", "Вівторок", "Середа", "Четвер", "П’ятниця"]
    for class_name in classes:
        print(f"\n\033[1;36m📘 Розклад для {class_name}\033[0m")
        table = []
        for day_index in range(5):
            row = [days[day_index]]
            for lesson_index in range(MAX_LESSONS_PER_DAY):
                lesson_key = f"Урок_{lesson_index + 1}"
                entry = schedule[class_name][day_index].get(lesson_key)
                if entry:
                    subject, teacher, room = entry
                    row.append(f"\033[92m{subject}\033[0m\n\033[90m({teacher}, {room})\033[0m")
                else:
                    row.append("-")
            table.append(row)
        headers = ["\033[1mДень\033[0m"] + [f"Урок {i + 1}" for i in range(MAX_LESSONS_PER_DAY)]
        print(tabulate(table, headers=headers, tablefmt="grid"))

# 🚀 === Запуск ===
best_schedule = genetic_algorithm(POPULATION_SIZE, NUM_GENERATIONS)
print_schedule(best_schedule)
```
