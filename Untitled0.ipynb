```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Набір даних із завдання 3
data_task3 = [20, 22, 25, 21, 19, 27, 30, 18, 26, 24, 29, 28, 23, 22, 25, 21, 20, 24, 26]

# Обчислення квартилів
Q1 = np.percentile(data_task3, 25)  # Перший квартиль (Q1)
Q2 = np.median(data_task3)          # Медіана (Q2)
Q3 = np.percentile(data_task3, 75)  # Третій квартиль (Q3)

# Обчислення IQR
IQR = Q3 - Q1

# Межі "вусів"
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

# Мінімальне та максимальне значення у межах вусів
min_val = min([x for x in data_task3 if x >= lower_whisker])
max_val = max([x for x in data_task3 if x <= upper_whisker])

# Побудова коробчастої діаграми
plt.figure(figsize=(6, 4))
sns.boxplot(x=data_task3, color='orange')
plt.title("Коробчаста діаграма (Boxplot)")
plt.xlabel("Значення")
plt.grid(True)
plt.show()

# Виведення результатів
print(f"Q1: {Q1}")
print(f"Медіана (Q2): {Q2}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Нижній вус: {min_val}")
print(f"Верхній вус: {max_val}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Набір даних
data_boxplot_outliers = [5, 7, 9, 10, 12, 15, 20, 21, 22, 25, 30, 35, 100]

# Обчислення квартилів
Q1 = np.percentile(data_boxplot_outliers, 25)  # Перший квартиль (Q1)
Q2 = np.median(data_boxplot_outliers)          # Медіана (Q2)
Q3 = np.percentile(data_boxplot_outliers, 75)  # Третій квартиль (Q3)

# Обчислення IQR
IQR = Q3 - Q1

# Межі "вусів" у Boxplot
lower_whisker = Q1 - 1.5 * IQR
upper_whisker = Q3 + 1.5 * IQR

# Визначення викидів
outliers = [x for x in data_boxplot_outliers if x < lower_whisker or x > upper_whisker]

# Побудова boxplot
plt.figure(figsize=(6, 4))
ax = sns.boxplot(x=data_boxplot_outliers, color='orange')
plt.title("Коробчаста діаграма (Boxplot) з викидами")
plt.xlabel("Значення")
plt.grid(True)

# Додавання текстових підписів для критичних чисел
critical_values = {"Q1": Q1, "Медіана": Q2, "Q3": Q3, "Нижній вус": lower_whisker, "Верхній вус": upper_whisker}
for label, value in critical_values.items():
    plt.text(value, 0.02, f"{label}\n{value:.1f}", horizontalalignment='center', verticalalignment='bottom', fontsize=10, color='red')

# Додавання підписів для викидів
for outlier in outliers:
    plt.text(outlier, 0.02, f"Викид\n{outlier}", horizontalalignment='center', verticalalignment='bottom', fontsize=10, color='blue')

plt.show()

# Виведення результатів
print(f"Q1: {Q1}")
print(f"Медіана (Q2): {Q2}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Нижній вус: {lower_whisker}")
print(f"Верхній вус: {upper_whisker}")
print(f"Викиди: {outliers}")
```
