```python
import numpy as np
import matplotlib.pyplot as plt
import csv

from scipy.integrate import quad
def f(x):
  return 5*np.sin(np.pi*5*x)


def b_k_quad(k):

    integrand = lambda x: f(x) * np.sin(k*x)
    result, error = quad(integrand, 0, np.pi)
    return (2/np.pi) * result
def S_N(x, N):

    result = 0
    for k in range(1, N+1):
        result += b_k_quad(k) * np.sin(k*x)
    return result
def save_results(x,N,  filename="results.csv"):

    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)

        # 1)  N
        writer.writerow([" N=", N])

        # 2)  b_k
        writer.writerow([])  #    
        writer.writerow(["k", "b_k"])
        for k in range(1, N+1):
          writer.writerow([k, b_k_quad(k)])


        # 3) 
        writer.writerow([])
        writer.writerow([" error", err(x,N)])
def err(x,N):
  return (f(x)-S_N(x,N))/f(x)


def graf(N):
  x = np.linspace(0, 0.2, 500)
  y_true = f(x)
  y_approx = [S_N(xi, N) for xi in x]

  plt.plot(x, y_true, label="f(x)", linewidth=2)
  plt.plot(x, y_approx, label="S_N(x), N=30", linestyle="--")
  plt.legend()
  plt.grid()
  plt.show()


def graf_bk(N):
  ks = np.arange(1, N+1)
  bs = [b_k_quad(k) for k in ks]

  plt.stem(ks, bs, basefmt=" ", markerfmt="o", linefmt="-")
  plt.xlabel("k")
  plt.ylabel("b_k")
  plt.title("  b_k")
  plt.grid()
  plt.show()

N=30
x=np.pi/2
print ("f(x)=",f(x))
print("Suma=",S_N(x, N))

print("Error=",(f(x)-S_N(x,N))/f(x))

graf(N)
graf_bk(N)
save_results(x,N,  filename="results.csv")

```
